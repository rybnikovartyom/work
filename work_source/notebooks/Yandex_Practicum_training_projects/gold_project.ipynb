### Проверка формулы рассчёта коэффициента эффективности обогощения Recovery

В описании технологического процесса мы получили формулу для рассчёта коэффициента эффективности обогощения.

Значение данного коэффициента приведено в полном и обучающем датасете, для проверки соответствия имеющихся значений формуле рассчитаем коэффициент обогащения на тренировочной выборке для этапа флотации - `rougher.output.recovery` и рассчитаем `MAE` между полученными и имеющимися результатами.

Обозначим переменные из формулы имеющимися в датасете полями:
* `C` - `rougher.output.concentrate_au` - доля золота в концентрате после флотации
* `F` - `rougher.input.feed_au` - доля золота в веществе до флотации
* `T` - `rougher.output.tail_au` - доля золота в отвальных хвостах после флотации

```python
recovery_calc = (data_dict['train']['rougher.output.concentrate_au'] * 
                     (data_dict['train']['rougher.input.feed_au'] - data_dict['train']['rougher.output.tail_au'])
                ) / \
                (data_dict['train']['rougher.input.feed_au'] * 
                     (data_dict['train']['rougher.output.concentrate_au'] - data_dict['train']['rougher.output.tail_au'])
                ) * 100

mae = mean_absolute_error(data_dict['train']['rougher.output.recovery'], recovery_calc)
print('Значение метрики MAE для рассчитанного и данного коэффициента', mae)
```

Как мы видим значение `MAE` между рассчитанным коэффициентом эффективности обогащения и данным в обучающем датасете для этапа флотации стремится к 0, имеющаяся разница скорее обусловлена округлениями нежели значимыми различиями результатов.

Также сразу проведём преобразование типа данных столбца `date` к типу `datetime`, далее по этому же столбцу проверим датасет на наличие дубликатов, т.к. судя по описанию, данное поле является своего рода ключом, и наличие дублей в этом поле будет говорить о необходимости объединения строк.

Заодно проверим датасеты на общее количество строк с пропусками хотя в одном поле.

```python
for name, df in data_dict.items():
    print('—'*80, '\nДатасет ', name,'\n','—'*80 ,sep='')
    df.rename(columns={x:x.replace('.', '_') for x in list(df.columns)}, inplace=True)
    df['date'] = pd.to_datetime(df['date'])
    print('Check 5 first renamed columns:', df.columns.to_list()[:5])
    print('Check date datatype:', df.date.dtype)
    print('Check duplicated values in date column:', df['date'].duplicated().sum())
    print('Rows with missing values:', df.isna().any(1).sum(), 'from', len(df))
```

### Изучение концентрации металлов на различных этапах очистки

Для изучения концентрации металлов (золота, серебра и свинца) на различных этапах очистки, для каждого металла построим по 2 графика:
* гистограмму распределения концентрации металлов по этапам очистки;
* линейный график изменения средней концентрации металла от этапа к этапу с 95% доверительным интервалом.

```python
stage = ['rougher_input', 'rougher_output', 'primary_cleaner_output', 'final_output']
state = ['feed', 'concentrate']
metals = {'Aurum': 'au', 'Argentum': 'ag', 'Plumbum': 'pb'}
substances = ['au', 'ag', 'pb', 'sol']

conc_dict = {met: [name for name in [stg+'_'+stt+'_'+m for stt in state for stg in stage] 
               if name in data_dict['full'].columns.to_list()]
     for met, m in metals.items()}

def concentration(data, metal, colomns):
    df_distr = pd.concat([pd.DataFrame(data[col].to_list(),
                                 index=[col]*len(data), 
                                 columns=['concentrate']) for col in colomns])\
            .reset_index().rename(columns={'index':'stage'})
    df_stat = pd.concat([data[colomns].mean(),data[colomns].quantile(0.025), data[colomns].quantile(0.975)], axis=1)
    sns.set_style("darkgrid")
    fig, ax0 = plt.subplots(figsize = (15,7))
    ax0 = sns.histplot(x=df_distr.concentrate,hue=df_distr.stage,bins=BINS,alpha=0.3, element='step', linewidth=2)
    #ax0.legend(colomns, loc=9, ncol=4, bbox_to_anchor=(0.5,1.03))
    ax0.set_title(f'Распределение концентрации {metal} по этапам', fontsize = 23, pad=20)
    fig, ax1 = plt.subplots(figsize = (15,7))
    ax1 = sns.lineplot(data=df_stat, markers=True, linewidth = 2 ,markersize=10)
    ax1.legend(['mean', 'lower', 'upper'])
    ax1.set_ylabel('concentration')
    ax1.set_xlabel('stages')
    ax1.set_title(f'Среднее и доверительный 95% интрервал концентрации {metal} по этапам', fontsize = 23);
```

```python
for met, cols in conc_dict.items():
    concentration(data_dict['full'], met, cols)
```

## Модель

Напишем функцию расчёта метрики sMAPE для целевого признака, а также функцию итогового sMAPE.

```python
def smape_func(answers,predictions):
    predictions = pd.Series(predictions, index=answers.index.to_list())
    return (abs(answers - predictions) / ((abs(answers) + abs(predictions)) / 2)).mean() * 100
 
smape_scorer = make_scorer(smape_func, greater_is_better=False)

def smape_agg_func(smape_rougher, smape_final):
    return smape_rougher * 0.25 + smape_final * 0.75
```

Разделим тренировочный и тестовый датасеты на признаки и целевые признаки.

```python
features_train = data_dict['train'][features_columns]
target_rougher_train = data_dict['train']['rougher_output_recovery']
target_final_train = data_dict['train']['final_output_recovery']

features_test = data_dict['test'][features_columns]
target_rougher_test = data_dict['test']['rougher_output_recovery']
target_final_test = data_dict['test']['final_output_recovery']
```

### Наивная модель DummyRegressor

В первую очередь создадим наивную модель с помощью `DummyRegressor` и на с помощью кросс-валидации оценим итоговую sMAPE, полученное значение будем использовать в дальнейшем для проверки построенных моделей на адекватность.

```python
estimator = DummyRegressor()
params={}
search_lin = GridSearchCV(estimator,params,cv=3, scoring=smape_scorer)

search_lin.fit(features_train,target_rougher_train)
smape_r = abs(search_lin.best_score_)
model_r = search_lin.best_estimator_

search_lin.fit(features_train,target_final_train)
smape_f = abs(search_lin.best_score_)
model_f = search_lin.best_estimator_

smape = smape_agg_func(smape_r,smape_f)
print(f'Итоговая sMAPE для наивной модели по среднему составляет {smape}')
```

Итак, в случае, если мы будем получать на построенных моделях значение итоговой sMAPE больше 10.38, то будем считать, что такая модель не проходит проверку на адекватность.

### Модель линейной регрессии

С помощью кросс-валидации оценим модель линейной регрессии на основании метрики итоговой sMAPE.

```python
estimator = LinearRegression()
params={}
search_lin = GridSearchCV(estimator,params,cv=3, scoring=smape_scorer)

search_lin.fit(features_train,target_rougher_train)
smape_r = abs(search_lin.best_score_)
model_r_lin = search_lin.best_estimator_

search_lin.fit(features_train,target_final_train)
smape_f = abs(search_lin.best_score_)
model_f_lin = search_lin.best_estimator_

smape = smape_agg_func(smape_r,smape_f)
print(f'Итоговая sMAPE для модели линейной регрессии составляет {smape}')
```

Сравнивая полученное значение метрики `sMAPE`, полученное с помощью линейной регрессии, с метрикой, полученной с помощью наивной модели, делаем вывод, что линейная модель не проходит проверку на адекватность.

Итоговая `sMAPE` по предсказаниям на тестовой выборке получилась даже чуть лучше, чем полученная на обучающей выборке с помощью кросс-валидации, и составила 9.41.


*Таким образом, мы можем рекомендовать полученную модель случайного леса к использованию для предсказания коэффициента эффективности восстановления золота.*
