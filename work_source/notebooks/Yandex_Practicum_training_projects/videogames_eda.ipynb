О качестве данных ничего не известно. Поэтому перед проведением исследования потребуется обзор данных, в результате которого мы сделаем первые выводы и предположения о качестве и структуре данных.

Следующим этапом мы приступим к исследовательскому анализу, в ходе которого и попытаемся установить основные взаимосвязи, выявить факторы и закономерности, влияющие на продажи видеоигр.

В конце используя статистические методы, попытаемся проверить следующие гипотезы:

Таким образом, исследование пройдёт в несколько этапов:
 * Обзор данных.

## Предобработка данных

Можно предположить, что оценки и рейтинг не подтянулись по каким-либо техническим причинам, нам же нужно принять решение, как обрабатывать эти пропуски.

Однако, перед этим вспомним и обратим внимание на то, что тип данных столбца `user_score` отображается как `object`. Посмотрим поближе на уникальные значения, которые присутствуют в этом столбце.



```python
data.user_score.unique()
```

Итак, мы видим, что в поле `user_score` присутствует текстовое значение - "tbd".

Судя по всему "tbd" является сокращением от "To Be Determined", т.е. "будет определено", это пометка в данных о том, что оценка пользователей ожидается, но ещё не получена.

Мы не можем заменить данный артефакт каким-либо числовым значением, т.к. любая оценка, даже 0, в данном случае имеет свою трактовку и может в последствии исказить данные. В связи с этим заменим значения "tbd" в поле `user_score` на `nan` и преобразуем тип данных этого поля в числовой.

```python
data['user_score'].where(data['user_score'] != 'tbd', np.nan, inplace=True)
data.user_score = pd.to_numeric(data.user_score)
```

### Проверка на наличие дубликатов

Проверим датафрейм на наличие полных дубликатов строк.

```python
data.duplicated().sum()
```

Полные дубликаты отсутствуют, это хорошо!

Но стоит проверить датасет на наличие дубликатов по столбцам за исключением столбцов с выручкой, т.к. если такие имеются, то с целью аггрегирования данных мы можем объединить такие строки просуммировав выручку по строкам.

```python
# список полей, по которым будем группировать
col_list = ['name', 'platform', 'year_of_release', 'genre', 'critic_score', 'user_score', 'rating']
# группируем применяя сумму ко всем полям с выручкой и count к полю с na_sales для поиска дублей
cat_col_grouped = data.groupby(col_list).agg(na_count = ('na_sales', 'count'),
                                            na_sales_sum = ('na_sales', 'sum'),
                                            eu_sales_sum = ('eu_sales', 'sum'),
                                            jp_sales_sum = ('jp_sales', 'sum'),
                                            other_sales_sum = ('other_sales', 'sum'))
# добавляем рассчитанные суммы для строк с количеством повторов больше 1 (дубли) в исходный датафрейм
data = data.merge(cat_col_grouped[cat_col_grouped['na_count'] > 1], on = col_list, how='left')
# заменяем значения выручки для дублящихся строк на суммы значений
data['na_sales'] = data['na_sales'].where(data['na_sales_sum'].isna(),data['na_sales_sum'])
data['eu_sales'] = data['eu_sales'].where(data['eu_sales_sum'].isna(),data['eu_sales_sum'])
data['jp_sales'] = data['jp_sales'].where(data['jp_sales_sum'].isna(),data['jp_sales_sum'])
data['other_sales'] = data['other_sales'].where(data['other_sales_sum'].isna(),data['other_sales_sum'])
# удаляем добавленные ранее вспомогательные колонки
data.drop(columns=['na_count', 'na_sales_sum', 'eu_sales_sum', 'jp_sales_sum', 'other_sales_sum'], inplace=True)
```

И снова проверим наличие дубликатов в датафрейме

```python
data.duplicated().sum()
```

С помощью наших преобразований мы нашли неявные дубликаты и теперь можем их удалить.

```python
data = data.drop_duplicates()
data.reset_index(drop = True, inplace=True)
```

Посчитаем медиану и среднее для платформ, у которых максимальным годом продаж не является 2016, т.к. это просто крайняя точка наблюдений, которая не обязательно характеризует окончание продаж, мы же в свою очередь хотим посмотреть именно "продолжительность жизни" платформы, от старта продаж до окончания.

## Выдвижение и статистическая проверка гипотез

В качестве проведения статистического исследования проверим два утверждения:
* Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
* Средние пользовательские рейтинги жанров Action и Sports разные.

Для первого утверждения сформулируем следующие основную и альтернативную гипотезы:

* **H₀:** Средние пользовательские рейтинги платформ Xbox One и PC одинаковы;
* **H₁:** Средние пользовательские рейтинги платформ Xbox One и PC различны

Для проверки гипотез будем использовать специальный метод для проверки гипотезы о равенстве среднего двух генеральных совокупностей по взятым из них выборкам: `scipy.stats.ttest_ind (array1, array2, equal_var)`.

Параметр `alpha` определим на уровне 0.05, для также будем использовать параметр `equal_var = False`, т.к. можно допустить, что дисперсия оценок пользователей для игр, выпущенных на различных платформах, может отличаться.

```python
pc_user_score = list(data_filtered.query('platform == "pc" and user_score.notna()')['user_score'])
xone_user_score = list(data_filtered.query('platform == "xone"  and user_score.notna()')['user_score'])
print('Длина выборки игр на PC: ', len(pc_user_score), '\n', 'Длина выборки игр на Xbox One: ', len(xone_user_score),sep='')
results = st.ttest_ind (pc_user_score, xone_user_score,equal_var=False)
alpha = .05
print(results.pvalue)
if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

Значение p-value приблизительно 0,28, это говорит о том, вероятность получить различия в данных СЛУЧАЙНО равна почти 28%, т.е. в 28% случаев при получении различий в средних значениях пользовательских оценок платформ Xbox One и PC эти различия будут обусловленны именно СЛУЧАЙНОСТЬЮ, но не носить систематический характер, таким образом, при установленной границе `alpha=0.05`, мы не можем отвергнуть нулевую гипотезу о равенстве средних двух генеральных совокупностей.

Для второго утверждения сформулируем следующие основную и альтернативную гипотезы:

* **H₀:** Средние пользовательские рейтинги жанров Action и Sports одинаковы;
* **H₁:** Средние пользовательские рейтинги жанров Action и Sports различны

Параметры `alpha` и `equal_var` будем использовать те же, что и для проверки предыдущей пары гипотез.

```python
action_user_score = list(data.query('genre == "action" and user_score.notna()')['user_score'])
sports_user_score = list(data.query('genre == "sports" and user_score.notna()')['user_score'])
print('Длина выборки игр жанра Action: ', len(action_user_score),
      '\n', 'Длина выборки игр жанра Sports: ', len(sports_user_score),sep='')
results = st.ttest_ind (action_user_score, sports_user_score,equal_var=True)
alpha = .05
print(results.pvalue)
if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

Значение p-value приблизительно 0.06, это говорит о том, вероятность получить различия в данных СЛУЧАЙНО равна почти 6%, т.е. в 6% случаев при получении различий в средних значениях пользовательских оценок игр жанров Action и Sports эти различия будут обусловленны именно СЛУЧАЙНОСТЬЮ, но не носить систематический характер, таким образом, при установленной границе `alpha=0.05`, мы не можем отвергнуть нулевую гипотезу о равенстве средних двух генеральных совокупностей.

Также не стоит забывать о PC, ничего не предвещает полного отказа от персональных компьютеров в плане ировой платформы, разработчики и геймеры продолжают активно использовать ПК для игр, это платформа с самым длительным жизненным циклом, потребление компьютерных игр год от года более стабильно и не отличается такими резкими скачками и падениями, нежели консольные аналоги;
* оценив корреляцию между оценками пользователей и продажами мы пришли к выводу, что пользовательские оценки никак не влияют на продажи. Оценки критиков же, в свою очередь, оказывают пусть и слабое, но прямопропорциональное влияние на продажи, чем выше оценка, тем больше игра продается, но ещё раз отметим, что коэффициент корреляции в данном случае около 0,3-0,4 - это слабая связь, т.е. можно ориентироваться на оценки критиков, но далеко не во всех случаях.


**Итак, какие закономерности и факторы стоит учитывать при планировании маркетинговой политики на будущие периоды:**
1. В первую очередь стоит смотреть какие платформы сейчас актуальны и набирают обороты, как правило популярность, как с точки зрения разработчиков игр, так и с точки зрения пользователей, той или иной платформы циклична - постепенный рост, достижение пика, и постепенное затухание продаж.

2. Самые интересный с точки зрения формирования дохода жанр - *shooter* ("стрелялки"), при относительно небольшом количестве выпускаемых игр, этот жанр занимает второе место в топ-5 продаж в абсолютном выражении по всем рынкам. На первом же месте в топ-5 продаж игры жанра *action*, но этих игр выпускается также и самое большое количество, таким образом, возникает вероятность допустить ошибку и выбрать для продажи не сумую коммерчески успешную игру. Также наряду с уже обозначенными жанрами достаточно популярны игры жанров *role-playing* и *sports*, не стоит забывать про них при формировании базового ядра продаваемых игр.

3. Обязательно стоит учитывать портрет потенциального потребителя в зависимости от региона:
* портрет североамериканского и европейского геймера во многом схож, они предпочитают играть на настольных консолях от американских и японских разработчиков в экшны, шутеры и спортивные игры с достаточно высокими возрастными ограничениями(рейтинг ESRB *М*). 

Самые актуальные консоли - ps4 (22% в Америке и 32% в Европе), x360 (20% в Америке и 12% в Европе), xone (20% в Америке и 12% в Европе), ps3(16% и 19% в Америке и Европе соответственно), 3ds(по 8%).

Самые популярные жанры - Action (29-31%), Shooter (21-23%), Sports(14%), Role-playing (9-10%), Misc(5-6%).

Распределение продаж по рейтингам - M (38%), не определено (20%), E (20%), E10+ (12%), T (10%);
* характерный портрет же японского геймера категорически отличается от европейцев и американцев, там отдают предпочтение портативным приставкам и больше всего играют в ролевые и экшн игры преимущественно без определения возрастного рейтинга.

Самые актуальные консоли - 3ds (49%), ps3 (18%), psv (12%), ps4(8%), wiiu (по 8%).

Самые популярные жанры - Role-playing (36%), Action (27%), Misc (6%), Fighting (5%), Shooter(4%).

Распределение продаж по рейтингам - не определено (58%), T (15%), E (11%), M (10%), E10+ (4%);
* cтоит отметить также тот факт, что геймеры представленных регионов отдают предпочтение платформам отечественной разработки, так в Японии в топе продаж игр платформы исключительно японских разработчиков, в Америке же на первом месте по продажам стоят игры для приставки линейки *Xbox* от американской компании *Microsoft*.
* также нужно обрать внимание на ёмкость рынков, американский рынок является самым крупным по продажам игр, далее идёт европейский рынок и самые низкие показатели продаж относительно представленных на японском рынке

Выявленные закономерности рекомендуется использовать для прогнозирования продаж и определения маркетинговой политики на будущие периоды.
